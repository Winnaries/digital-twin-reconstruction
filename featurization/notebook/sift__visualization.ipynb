```python
import numpy as np
from numpy import asarray
from PIL import Image
import matplotlib.pyplot as plt
```

```python
im = Image.open("../demo/table.webp")
im = asarray(im)
im.shape
```

```python
keypoints = np.load("../demo/keypoints.npz")
X = keypoints[:, 0]
Y = keypoints[:, 1]
sigma = keypoints[:, 2]
plt.imshow(im)
plt.scatter(Y, X, s=sigma * 6, alpha=0.5)
```

```python
octave = 0
delta = 2 ** octave * 0.5

unfiltered = np.load("../demo/keypoints-unfiltered.npz")
O = unfiltered[:, 0]
X = unfiltered[:, 1][O == octave]
Y = unfiltered[:, 2][O == octave]
plt.imshow(im)
plt.scatter(Y * delta, X * delta, s=3, alpha=0.5)
```
